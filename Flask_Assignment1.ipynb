Q1. What is Flask framework? What are advantages of Flask framework?

Flask is framework in python which is used for web development. It is a lightweight framework which 
is also used in API development. It is a simple, easy and flexible framework which makes it popular 
amongst the developers.

Advantages of Flask are the following:
Lightweight: It has a relatively low code-base which makes helps itself for small or medium scale web 
development
Easy to use: Due to its intuitive API, it becomes a popular choice among developers to learn and use
Community Support: Flask community has a significant number of developers who support and contribute to the framework

Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is used for defining the routes for specific functions in web application.
 We use app routes in Flask for several reasons:

URL handling: App routes allow us to handle different URLs in a web application and direct them to the appropriate views or functions.

Navigation: App routes make it easy to navigate between different pages or views in a web application.

User input: App routes can be used to handle user input and process it accordingly. For example, when a user submits a form, the app route can handle the form data and perform the necessary actions.

RESTful API: App routes can be used to create RESTful APIs in Flask, which allow other applications to interact with the web application via HTTP requests.
