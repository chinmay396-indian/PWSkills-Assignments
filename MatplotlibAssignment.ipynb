Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library for Python. It is widely used for creating high-quality, customizable graphs, charts, and plots. It was developed by John D. Hunter in 2003 as an open-source project and has since become the standard data visualization library in Python.

Matplotlib is used for a variety of purposes, such as:

Exploring and analyzing data
Communicating insights and findings through visualizations
Creating publication-quality graphics for scientific papers and presentations
Developing interactive visualizations for web applications
The Pyplot module of Matplotlib provides a simple interface for creating a wide variety of plots, including:

Line plot - It is a plot in which the data points are connected by a line. It is used to visualize the trend in data over time.

Bar plot - It is a plot in which the data is represented by vertical or horizontal bars. It is used to compare the values of different categories.

Histogram - It is a plot that shows the frequency distribution of a variable. It is used to visualize the shape of the distribution of a variable.

Scatter plot - It is a plot in which the data points are represented by markers. It is used to visualize the relationship between two variables.

Heatmap - It is a plot in which the data is represented by color-coded cells. It is used to visualize the correlation between two or more variables.

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

The subplot() function is used to create multiple plots within a single figure. It allows you to arrange multiple plots in a grid, where you specify the number of rows and columns of the grid and the position of each plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# define the data for each line
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# create a figure with a 2x2 grid of plots
fig, axs = plt.subplots(2, 2)

# plot each line on a separate plot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# set the titles for each plot
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# add a common x-axis label and y-axis label
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')

# adjust the layout to avoid overlapping titles
plt.tight_layout()

# show the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.



A bar plot is a type of plot that represents categorical data with rectangular bars. The height or length of each bar corresponds to the value of the category it represents. Bar plots are used to visualize and compare the frequency, count, or proportion of different categories.

Bar plots are commonly used when:

The data is categorical, with a small number of categories.
You want to compare the frequency, count, or proportion of each category.
You want to emphasize the difference between categories.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel = "Company Name"
plt.ylabel = "Profit"
plt.title = "Profit Chart"
plt.show()
plt.barh(company,profit)
```

A box plot, also known as a box-and-whisker plot, is a type of graphical representation used to display the distribution of a dataset. It shows the median, quartiles, and outliers of the data in a concise way. The box represents the interquartile range (IQR), which is the range between the first and third quartiles (Q1 and Q3, respectively). The whiskers extend from the box to the minimum and maximum values within 1.5 times the IQR, and any data points outside this range are shown as individual points or outliers.

Box plots are commonly used in statistics and data analysis to visually summarize the distribution of a dataset, particularly when comparing multiple datasets. They can help identify differences in central tendency, variability, and skewness of the data.
