Q3. Why is redirect() used in Flask?

In Flask, a redirect is used to send a user to a different URL or route within the application. This is often used to redirect a user after a successful form submission, or to redirect a user to a login page if they are not authenticated.

Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are used to generate dynamic HTML pages that can be returned as responses to HTTP requests. Templates allow developers to separate the presentation logic of a web application from its application logic, making it easier to manage and modify the design of a website.

In Flask, templates are typically written in HTML, with placeholders for dynamic content that are filled in at runtime. Flask supports a number of templating engines, including Jinja2, which is the default templating engine for Flask.
